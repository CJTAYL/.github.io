# Stock Dashboard

```python
import plotly.express as px
import yfinance as yf

# Micron Technologies
mu = yf.download(tickers='MU', period='15d', interval='30m')
# Kinder Morgan Inc. 
kmi = yf.download(tickers='KMI', period='15d', interval='30m')
# Ares Capital 
arc = yf.download(tickers='ARCC', period='15d', interval='30m')
```

**Ares Capital Corporation** (ARCC)

A leading specialty finance company that provides one-stop solutions to meet the distinct and underserved financing needs of private middle-market companies across diverse industries (Ares Capital).

```python
for template in ["plotly_dark"]:
    fig = px.line(arc, y='Close', title='Ares Capital - Purchase Price: $19.15', template=template)
    fig.show()
```

**Kinder Morgan Inc. **(KMI)

One of the largest energy infrastructure companies in North America. The company specializes in owning and controlling oil and gas pipelines and terminals. Kinder Morgan owns an interest in or operates approximately 83,000 mi of pipelines and 143 terminals (Wikipedia).

```python
fig = px.line(kmi, y='Close', title='Kinder Morgan Inc. - Purchase Price: $17.93', template=template)
fig.show()
```
